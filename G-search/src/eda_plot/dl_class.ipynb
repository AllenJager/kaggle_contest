```python
#基于物品的协同过滤算法
import math
#数据集
Train_Data = {
     'A':{'i1':1,'i2':1 ,'i4':1},
     'B':{'i2':1,'i4':1},
     'C':{'i1':1,'i2':1,'i5':1},
     'D':{'i2':1,'i3':1}
     }
#计算物品相似度
#初始化一个空字典
C = dict()  ##物品对同时被购买的次数
N = dict()  ##物品被购买用户数
for u, items in Train_Data.items():  # train.items()是把train中key,value组成元组返回
    for i in items.keys():  # item是物品
        if i not in N.keys():
            N[i] = 0
        N[i] += 1
        for j in items.keys():
            if i == j:
                continue
            if i not in C.keys():
                C[i] = dict()    #字典C的值也是字典的形式
            if j not in C[i].keys():
                C[i][j] = 0
            ##当用户同时购买了i和j，则加1
            C[i][j] += 1
W = dict()  ##物品对相似分数
for i, related_items in C.items():
    if i not in W.keys():
        W[i] = dict()  #W的值也是字典
    for j, cij in related_items.items():
        W[i][j] = cij / math.sqrt(N[i] * N[j])

print(W)

#进行推荐
def Recommend(train,user_id,W,K):  #K超参数，需要调整，过大，有很多不相关的物品，过小，准确率下降
    rank = dict()
    ru = train[user_id] #给出user_id买过的商品，其值是字典型  #'C':{'i1':1,'i2':1,'i5':1},
    for i,pi in ru.items():
        tmp=W[i]
        for j,wj in sorted(tmp.items(),key=lambda d: d[1],reverse=True)[0:K]:  #选出最相似的的K个物品计算它们的分数
            if j not in rank.keys():
                rank[j]=0
            if j in ru:  #r如果用户已经购买过，则不再推荐
                continue
            rank[j] += pi*wj #待推荐的物品j与用户已购买的物品i相似，则累加上相似分数
    return rank

Rank=Recommend(Train_Data,'C',W,3)
print(Rank)
```
